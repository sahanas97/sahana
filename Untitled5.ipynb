```python
Q1 = df['horsepower'].quantile(0.25)
Q3 = df['horsepower'].quantile(0.75)
iqr = Q3 - Q1
```

```python
upper_limit = Q3 + 1.5 * iqr
lower_limit = Q1 - 1.5 * iqr
print(upper_limit)
print(lower_limit)
```

```python
#Finding
df[(df['horsepower'] > upper_limit) | (df['horsepower'] < lower_limit)]
```

```python
df[(df['horsepower'] < upper_limit) & (df['horsepower'] > lower_limit)]
```

```python
#replacing
import numpy as np
df['horsepower'] = np.where(
    df['horsepower'] > upper_limit,
    upper_limit,
    np.where(
    df['horsepower'] < lower_limit,
    lower_limit,
    df['horsepower']
    )
)
df
```

```python
sns.boxplot(df['horsepower'])
plt.show
```

```python
Q1=df['acceleration'].quantile(0.25)
Q3=df['acceleration'].quantile(0.75)
iqr=Q3-Q1
```

```python
#Finding
df[(df['acceleration'] > upper_limit) | (df['acceleration'] < lower_limit)]
```

```python
# non outliers
df[(df['acceleration'] < upper_limit) & (df['acceleration'] > lower_limit)]
```

```python
#replacing
df['acceleration'] = np.where(
    df['acceleration'] > upper_limit,
    upper_limit,
    np.where(
    df['acceleration'] < lower_limit,
    lower_limit,
    df['acceleration']
    )
)
df
```
