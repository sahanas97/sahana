```python
#creating sample data for plotting
x = np.arange(1,16)    #1 … 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
#creating an empty canvas/figure
fig = plt.figure(figsize=(8,5))
#setting axes as [left, bottom, width, height]
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.10,0.42,0.45,0.45])
#plotting the lines
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label
ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label
ax1.legend()    #adding legend
ax2.legend()
```

```python
#creating sample data for plotting
x = np.arange(1,16)    #1 … 15
y = x**2    #square of x values
z = (y/2)+40    #arithmetic operation applied to y values
fig = plt.figure(figsize=(8,9))
ax1 = fig.add_axes([0,0,0.5,0.5])
ax2 = fig.add_axes([0.6,0,0.5,0.5])
ax3 = fig.add_axes([0,0.6,0.5,0.5])
ax4 = fig.add_axes([0.6,0.6,0.5,0.5])


#initializing four axes 
fig, (ax1,ax2,ax3,ax4)   
ax1.plot(x,y,color="g", label='line1' )
ax1.set_title('Title for Plot 1',fontsize=15)    #setting title
ax1.set_xlabel('Plot 1 - X axis',fontsize=15)    # setting X label
ax1.set_ylabel('Plot 1 - Y axis',fontsize=15)    # setting Y _label

ax2.plot(x,z,color="r", label='line2')
ax2.set_title('Title for Plot 2',fontsize=15)    #setting title
ax2.set_xlabel('Plot 2 - X axis',fontsize=15)    # setting X label
ax2.set_ylabel('Plot 2 - Y axis',fontsize=15)    # setting Y _label

ax3.plot(x,z,color="b", label='line3')
ax3.set_title('Title for Plot 3',fontsize=15)    #setting title
ax3.set_xlabel('Plot 3 - X axis',fontsize=15)    # setting X label
ax3.set_ylabel('Plot 3 - Y axis',fontsize=15)

ax4.plot(x,z,color="y", label='line4')
ax4.set_title('Title for Plot 4',fontsize=15)    #setting title
ax4.set_xlabel('Plot 4 - X axis',fontsize=15)    # setting X label
ax4.set_ylabel('Plot 4 - Y axis',fontsize=15)
fig.tight_layout()
```

```python
import pandas as pd
df=pd.read_csv('dataset/auto_mpg.csv')
df
```

```python
#replacing the null values in 'horsepower' with its mean value
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)
```

```python
#visualizing correlation between mileage and horsepower
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[10,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
#plotting scatter plot
ax.scatter(df['horsepower'],df['mpg'])
ax.set_title("Scatter plot of horsepower and mileage",fontsize=16)
ax.set_ylabel('Mileage per gallon',fontsize=12)
ax.set_xlabel('horsepower',fontsize=12)
```

```python
#finding correlation between mileage and horsepower 
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='horsepower',y='mpg',hue='origin',data=df)
```

```python
df=pd.read_csv('dataset/auto_mpg.csv')
df.info()
```

```python
df['horsepower'].fillna(df['horsepower'].mean(),inplace=True)
```

```python
df['mpg'].unique()
```

```python
df['mpg'].describe()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,6])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Distribution of mileage')
ax.boxplot(df['mpg'])
fig.show()
```

```python
df['city_mileage']=df['mpg']-df['mpg']*0.25
df
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[6,5])
ax=fig.add_axes([0,0,1,1])
ax.set_title('Distribution of average MPG vs city MPG')
ax.set_ylabel('mileage per gallon')
ax.boxplot([df['mpg'],df['city_mileage']])
fig.show()
```

```python
#line chart
line_df=df.groupby(by=['model_year'])
line_df=line_df.mean()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[8,4])
ax=fig.add_axes([0,0,1,1])
ax.plot(line_df['mpg']/10,label='mileage/10')
ax.set_title('mileage',fontsize=17)
ax.set_ylabel('Value',fontsize=17)
ax.set_xlabel('Model year',fontsize=17)
ax.legend()
fig.show()
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=[8,4])
ax=fig.add_axes([0,0,1,1])
ax.plot(line_df['weight']/1000,label='weight/1000')
ax.set_title('weight',fontsize=17)
ax.set_ylabel('Value',fontsize=17)
ax.set_xlabel('Model year',fontsize=17)
ax.legend()
fig.show()
```
