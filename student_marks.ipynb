```python
df['placement_exam_marks_zscore']=(df['placement_exam_marks']-df['placement_exam_marks'].mean())/df['placement_exam_marks'].std()
```

```python
df[(df['placement_exam_marks_zscore']>3)|(df['placement_exam_marks_zscore']<-3)]
```

```python
new_df=df[(df['placement_exam_marks_zscore']<3)&(df['placement_exam_marks_zscore']>-3)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(3,4))
sns.boxplot(df['placement_exam_marks'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(3,4))
sns.distplot(df['placement_exam_marks'])
plt.show()
```

```python
Q1=df['placement_exam_marks'].quantile(0.25)
Q3=df['placement_exam_marks'].quantile(0.75)
iqr=Q3-Q1
```

```python
upper_limmit=Q3+1.5*iqr
lower_limmit=Q1-1.5*iqr
```

```python
#finding
df[(df['placement_exam_marks']>=upper_limmit)|(df['placement_exam_marks']<lower_limmit)]
```

```python
df['placement_exam_marks'][311]
```

```python
#removing
new_df=df[(df['placement_exam_marks']<=upper_limmit) & (df['placement_exam_marks']>lower_limmit)]
#new_df
plt.figure(figsize=(3,4))
sns.boxplot(new_df['placement_exam_marks'])
plt.show()
```

```python
import numpy as np
df['placement_exam_marks']=np.where(
df['placement_exam_marks']>=upper_limmit,
upper_limmit,
np.where(df['placement_exam_marks']<lower_limmit,
        lower_limmit,
        df['placement_exam_marks']
        ))
df
```
