```python
import pandas as pd
df=pd.read_excel('dataset/SaleData.xlsx')
df
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df
```

```python
df1=df.groupby('Region').sum()['Sale_amt']
#plotting the number of cars in each year
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[7,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("Total amount of sales",fontsize=16)
ax.set_ylabel('sale amount',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
#plotting bar graph
ax.bar(df1.index, df1)
fig.show()
```

```python
df1=df.groupby('Item').sum()['Units']
#plotting the number of cars in each year
#importing the required packages
import matplotlib.pyplot as plt
#creating an empty canvas/figure
fig = plt.figure(figsize=[7,5])
#setting axes
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title("items",fontsize=16)
ax.set_ylabel('units',fontsize=12)
ax.set_xlabel('Model year',fontsize=12)
#plotting bar graph
ax.bar(df1.index, df1)
fig.show()
```
